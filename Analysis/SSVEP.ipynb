```python
events,all_events_id=mne.events_from_annotations(raw,event_id='auto')
```

```python
for idx in ['BH', 'F', 'LH','RH', 'SSVEP']:
    print(idx + ' is ' + str(all_events_id[idx]))
```

```python
pick_ssvep =  mne.pick_channels(raw.info["ch_names"], include=["O1", "O2", "PO3", "PO4","P3","P4","Pz","CP1","CP2"])
```

```python
event_id = dict({'SSVEP': 12})
tmin = 1
tmax = 3
#ssvep_epochs = mne.Epochs(raw, events, event_id, 
#                          tmin, tmax,
#                          baseline=None , preload=True, event_repeated = 'drop')
ssvep_epochs = mne.Epochs(raw, events, event_id, 
                          tmin, tmax, picks = pick_ssvep,
                          baseline=None , preload=True, event_repeated = 'drop')
```

```python
ssvep_epochs.info
```

```python
list_freqs = (6,7.5,10,15)
fs = 125
preds = []
for i in range(len(ssvep_epochs.events[:,2])):
    pred = fbcca.fbcca_realtime(ssvep_epochs.get_data(copy=True)[i], list_freqs, fs, num_harms=6, num_fbs=4)
    if pred == 0 or pred == 4:
         preds.append(1)
    else:
         preds.append(0)
print(np.mean(preds))
```

```python
event_id = dict({'BH': 1,
                 'F': 8,
                 'LH': 9,
                 'RH': 10,
                 'SSVEP':12})
tmin = 0
tmax = 3
epochs = mne.Epochs(raw, events, event_id, 
                    tmin, tmax, picks = pick_ssvep,
                    baseline=None , preload=True, event_repeated = 'drop')
```

```python
epochs.events[:,2]
```

```python
list_freqs = (6,7,8,9)
fs = 125
preds = []
for i in range(len(epochs.events[:,2])):
    pred = fbcca.fbcca_realtime(epochs.get_data(copy=True)[i], list_freqs, fs, num_harms=5, num_fbs=4)
    if epochs.events[:,2][0] == 12:    
        if pred == 1:
             preds.append(1)
        else:
             preds.append(0)
    else:
        if pred != 0:
             preds.append(1)
        else:
             preds.append(0)
print(np.mean(preds))
```
